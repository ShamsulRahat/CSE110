### Task 3

Write a python program that reads 5 numbers from the user into a list, and then prints them in the reverse order.

*Hint: You may create a list to store the input numbers and then use loop to print them in reverse order*

===================================================================

**Sample Input:**\
5\
-5\
100\
1\
0

**Sample Output:**\
0\
1\
100\
-5\
5

===================================================================

```python
#to do
user_input=[]
n=int(input('list numbers: '))
for i in range(n):
    user_input.append(int(input('numbers: ')))
print(user_input)
end= len(user_input)-1
ans=[]

while end>=0:
    number=user_input[end]
    ans.append(number)
    end=end-1
print(ans)
```
